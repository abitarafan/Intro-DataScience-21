**Gaussian Naive Bayesian Algorithm**

```python
import pandas as pd
from sklearn.naive_bayes import GaussianNB
from sklearn import model_selection
from sklearn.model_selection import train_test_split
from sklearn.metrics import classification_report
```

```python
# Loading students dataset
df = pd.read_csv("student-data.csv")
df.head()
```

```python
df.isnull().sum()
```

```python
df.describe()
```

```python
df.shape
```

```python
clf = GaussianNB()

array = df.values
x = array[:,1:47]
y = array[:,47]
```

```python
y = y.astype('int')
```

```python
print("input=", x)
print("output= ", y)
```

```python
# Validation dataset

validation_size = 0.33
```

```python
seed = 6
X_train, X_validation, Y_train, Y_validation = model_selection.train_test_split(x, y, test_size=validation_size, random_state=seed)

# Test evaluation metric
scoring = 'accuracy'
```

```python
#Fitting training set
clf.fit(X_train, Y_train)

#Predicting Test set
pred_clf = clf.predict(X_validation)

#Prediction Probability
prob_pos_clf = clf.predict_proba(X_validation)[:, 1]
```

```python
#Reshapig prediction dataframe to insert prediction results
pred_clf_df = pd.DataFrame(pred_clf.reshape(36,1))
pred_clf_df.rename(columns={0:'Grade'}, inplace=True)

#Reshaping the test dataset
X_validation_df = pd.DataFrame(X_validation.reshape(36,46))

```

```python
#Concatenating the two pandas DataFrames over the columns to create a prediction dataset
pred_outcome = pd.concat([X_validation_df, pred_clf_df], axis=1)
pred_outcome = pred_outcome.reindex(X_validation_df.index)

```

```python
pred_outcome.set_axis(df.columns, axis=1, inplace=True)
```

```python
del df['ID']
```

```python
#Merging the prediction with original dataset
pred_comp = pd.merge(df,pred_outcome)
```

```python
print((pred_comp).head(10))

print("\n")

#Save the file to csv
pred_comp.to_csv('Predictions.csv', sep=',')


#Save the file to Excel
from pandas import ExcelWriter

writer = ExcelWriter('GradesPredictions.xlsx')
pred_comp.to_excel(writer,'Sheet1')
writer.save()
```

```python
#Model Performance
#setting performance parameters
kfold = model_selection.KFold(n_splits=10)

#calling the cross validation function
cv_results = model_selection.cross_val_score(GaussianNB(), X_train, Y_train, cv=kfold, scoring=scoring)

#displaying the mean and standard deviation of the prediction
msg = "%s: %f (%f)" % ('NB accuracy', cv_results.mean(), cv_results.std())
print(msg)
```

**Random Forest Classifier**

```python
from sklearn.ensemble import RandomForestClassifier
import numpy as np
```

```python
np.random.seed(0)
```

```python
df.head()
```

```python
df['is_train'] = np.random.uniform(0, 1, len(df)) <= .75
```

```python
df.head()
```

```python
# Create two new dataframes, one with the training rows, one with the test rows
train, test = df[df['is_train']==True], df[df['is_train']==False]
```

```python
# Show the number of observations for the test and training dataframes
print('Number of observations in the training data:', len(train))
print('Number of observations in the test data:',len(test))
```

```python
# Create a list of the feature column's names
features = df.columns[:-2]

# View features
features
```

```python
y = pd.factorize(train['Grade'])[0]

y
```

```python
# Create a random forest Classifier. By convention, clf means 'Classifier'
clf = RandomForestClassifier(n_jobs=2, random_state=0)

# Train the Classifier to take the training features and learn how they relate
# to the training y (the species)
clf.fit(train[features], y)
```

```python
# Apply the Classifier we trained to the test data (which, remember, it has never seen before)
clf.predict(test[features])
```

```python
# View the predicted probabilities of the first 10 observations
clf.predict_proba(test[features])[0:10]
```

```python
test['Grade'].head()
```

```python
preds = clf.predict(test[features])
```

```python
pd.crosstab(test['Grade'], preds, rownames=['Grade'], colnames=['Predicted Grade'])
```
